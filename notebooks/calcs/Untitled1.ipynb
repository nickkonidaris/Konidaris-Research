```python
from pylab import *
import numpy as np
```

```python
N = 1200
no = 150
x = np.linspace(0,np.pi*no,N)
y = np.sin(x)
IMG = np.repeat(y, N)
IMG.resize(N,N)
```

```python
figure(figsize=(50,50))
imshow(IMG, cmap="gray")
```

```python
F = np.fft.fft(IMG)
```

```python
imshow(np.abs(F))
```
