```python
import numpy as np
from pylab import *

from numpy.polynomial import Chebyshev

%matplotlib inline
```

```python
t=np.array([93,113,133,153,173,193,213,233,253,273,293])
c=np.array([6.7,9.1,11.1,12.8,14.1,15.2,16.2,17,17.7,18.3,18.7])
```

```python
fit = Chebyshev.fit(t,c, deg=4)

plot(t,c,'o')
plot(t,fit(t))
```

```python
plot(t,c-fit(t),'o')
```

```python
I = fit.integ()
I(120)-I(300)
```
