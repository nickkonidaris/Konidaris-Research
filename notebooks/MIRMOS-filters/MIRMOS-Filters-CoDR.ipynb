```python
import numpy as np
from pylab import *



%matplotlib inline
```

```python
ii = np.loadtxt("transmission.dat.txt")
ll, T = ii.T
ll /= 1000
```

```python
rcParams['font.size'] =  22

hfnt = {} # {'fontname':'Arial'}


b = 'black'
figure(figsize=(11,10/1.61))
w = .2
plot(ll, T, color=b, lw=w, **hfnt)



curves = [0.7, .8860, 1.1240, 1.3520, 1.4660, 1.807, 1.9210, 2.404]
for c in curves: axvline(c, color=b, lw=3, **hfnt)
    
xlabel("Wavelength [µm]", **hfnt)
ylabel("Transmission", **hfnt)

ylim(0,1.2)

axhline(1, color=b, lw=.5)


v = 1.1
text(.78, v, "i")
text(0.98, v, "Y")
text(1.22, v, "J")
text(1.6, v, "H")
text(2.15, v, "K")

xlim(.6, 2.5)


```
