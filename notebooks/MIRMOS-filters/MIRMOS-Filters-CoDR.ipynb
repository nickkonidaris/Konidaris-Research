```python
import numpy as np
from pylab import *



%matplotlib inline
```

```python
i1 = np.loadtxt("atran.smo.22125.webarchive", encoding = 'utf-8-sig', unpack=True)
i2 = np.loadtxt("atran.smo.22178.webarchive", encoding = 'utf-8-sig', unpack=True)
```

```python
_, l1, T1 = i1
_, l2, T2 = i2
```

```python
rcParams['font.size'] =  22

hfnt = {} # {'fontname':'Arial'}


b = 'black'
figure(figsize=(10,10/1.61))
w = .2
plot(l1,T1, color=b, lw=w, **hfnt)
plot(l2,T2, color=b, lw=w, **hfnt)


curves = [.8860, 1.1240, 1.3520, 1.4660, 1.807, 1.9210, 2.404]
for c in curves: axvline(c, color=b, lw=3, **hfnt)
    
xlabel("Wavelength [m]", **hfnt)
ylabel("Transmission", **hfnt)

ylim(0,1.2)

axhline(1, color=b)


v = 1.1
text(0.98, v, "Y")
text(1.22, v, "J")
text(1.6, v, "H")
text(2.2, v, "K")



```
