## Scintillation Noise

```python
def young_scintillation_noise(obs, verbose=False):
    """ Returns the fractional /f """

    D = obs[Dtel]
    X = obs[airmass]
    h0 = 8000*u.m
    h_lco = 2000*u.m
    
    
    Dravins__f = np.array(0.09) * np.power(D/u.cm, -2/3) * np.power(X, 1.75) * np.exp(-h_lco/h0)/np.sqrt(2*obs["et"]/u.s)
    
    # Kornilov et al. :
    Kornilov__f = 0.0030 * (D/u.m)**(-2/3) * X**1.75 * np.exp(-h_lco/h0)/np.sqrt(obs["et"]/u.s)
    
    # Jason's Mystery paper:
    h0 = 5500*u.m

    m_s_f = 1.6 * (D/u.m)**(-2/3) * X**1.75 * (8000/5500)**(-7/12) / np.sqrt(obs["et"]/u.s)
    
    # Osborn Variance
    Osborn_var = 10 * 1e-6 * (D/u.m)**(-4/3) / (obs["et"]/u.s) * X**(-3) * np.exp(-2*h_lco/h0)
    
    if verbose: print(Dravins__f.cgs, Kornilov__f.cgs, m_s_f.cgs, np.sqrt(Osborn_var.cgs))
    
    return Dravins__f.cgs * 1.3
```

## Signal to Noise Ratios

```python
def snr_per_exp(targ_mag, obs, thpt=0.35, eff=15000*u.Angstrom, well_sig=60000):
    
    targ_fd = targ_mag.to(u.erg/u.s/u.Angstrom/u.cm**2, u.spectral_density(eff))
    
    obs["flux_density"] = targ_fd
    
    hc=h*c
    epp = hc/eff
     = eff/obs[Res]
    Atel = np.pi * (obs[Dtel]/2)**2
    targ_eps  = (thpt * targ_fd *  * Atel / epp)
    tar_eps_pix = targ_eps / obs[res_size_pix]
    
    tar_eps_pix = (60000.0/tar_eps_pix).cgs
    
    et = (clamp(20*u.s, tar_eps_pix, 240*u.s)).cgs
    
    obs["N/s"] = targ_eps.cgs
    obs["N_res"] = (targ_eps*et).cgs
    obs["et"] = et
    eff = (et/(et+obs["readout_overhead"])).cgs
    obs["observing_efficiency"] = eff
    obs["N_pix"] = obs["N_res"] / obs[res_size_pix] * eff
    
    f = young_scintillation_noise(obs)

    obs["scintillation_f_per_exp"] = f

    # binning factor
    bf = obs[Res]/obs["Res_Binned"]
    obs["N_res"] = obs["N_res"] * bf
    
    
    rn = np.sqrt(bf*obs["res_size_pix"])*obs[rn_pix]
    #var_res = rn**2 + bf*obs["N_pix"] + (obs["N_res"] * f)**2
    obs["rn_res"] = rn
    obs["shotnoise_res"] = np.sqrt(obs["N_res"])
    obs["scintnoise_res"] = obs["N_res"] * f
    
    var_res = obs["rn_res"]**2 + obs["shotnoise_res"]**2 + obs["scintnoise_res"]**2
    snr_res = obs["N_res"] / np.sqrt(var_res)
    
    
    obs["snr_res"] = snr_res.cgs

    obs["noise_res"] = np.sqrt(var_res)
    obs["DQE_res"] = snr_res/np.sqrt(obs["N_res"])
    
    return obs

def snr_per_transit(targ_mag, total_time, obs):
    
    single_exp = snr_per_exp(targ_mag, obs)
    
    Nexp_needed = np.ceil((total_time / single_exp["et"]).cgs)
    
    obs["Nexp_needed"] = Nexp_needed
    obs["snr_res_sum"] = obs["snr_res"] * np.sqrt(Nexp_needed)
    obs["unc_res_sum"] = 1/obs["snr_res_sum"]
    
    # Noise Propagation
    def f(x): obs[x+"_sum"] = obs[x] / np.sqrt(Nexp_needed)
    f("rn_res") ; f("shotnoise_res") ; f("noise_res") ; f("scintnoise_res")
    
    
    obs["N_sum"] = obs["N_res"] * Nexp_needed
    obs["DQE_sum"] = obs["snr_res_sum"] / np.sqrt(obs["N_sum"])
    
    return obs
```

# Slit Width

```python
from astropy.modeling.models import Moffat1D
```

```python

s1 = Moffat1D()

x = np.arange(-100,100,.1)
plot(x,s1(x))
```

```python
s1.alpha=1
```

```python
s1.fwhm
```

```python
in_slit = (x>-5) & (x<5)
shifted = (x>-5.3) & (x<4.7)
```

```python
y = s1(x)
```

```python
total_in = sum(y[in_slit])
total = sum(y)

centered = total_in/total
offset = sum(y[shifted])

print((centered-offset/total)/1e-6)
```
