# What is the infrared sky background?

Here we use the data from Oliva+ (2015) to compute the inter-line sky background.


```python
import numpy as np
import os
%matplotlib inline
from pylab import *

import astropy
from astropy.convolution import convolve
```

Read in the night sky spectrum from Oliva 2015

```python
dat = np.loadtxt("data/Oliva2015/table4.dat")
ll,ss = dat.T
```

Filter curves from MOSFIRE bandpass definitions

```python
Y = "Y" ; J = "J" ; H = "H" 

filters = {Y: (.972, 1.124), J: (1.153, 1.352), H: (1.466, 1.807)}
```

Make the luminosity specific in wavelength

## compare to paper


![alt text](fig3.png "Title")


```python
fig = figure(figsize=(16,6))
ax = fig.add_subplot(111)   
sp = subplot(1,1,1)


locs = [(15180, 15290), (15145, 15195), (15968, 16035), (16750, 16845)]

for ix, loc in enumerate(locs):
    subplot(2,2,ix+1)
    step(ll, ss, 'k')
    axhline(0, color='r', lw=.5)
    xlim(*loc)
    ylim(-.2,2.2)
```

# Plot of the spectrum

```python


for band, hpps in filters.items():
    
    
    l, r = hpps[0] * 10000, hpps[1] * 10000
    
    figure(figsize=(15,4))
    ok = (ll > l) & (ll < r)
    lc, sc = ll[ok], ss[ok] * Atel * Asky * et
    plot(lc, sc,'k-')
    grid(True)

    title("%s band" % band)
    xlabel("Wavelength [$\AA$]")
    ylabel("Flux [$photons$ for Magellan in a %s s exposure]" % et)

    axvline(l)
    axvline(r)
    #axvline(hpps[1]*10000)
```

# Now convolve down the spectrum

```python


for band, hpps in filters.items():
    
    l, r = hpps[0] * 10000, hpps[1] * 10000

    ok = (ll > l) & (ll < r)
    lc, sc = ll[ok], ss[ok] * Atel * Asky * et
    
    spectrum_dl = ll[1] - ll[0]
    Res = 4000
    desired_dl = np.median(ll) / Res
    Npix = np.round(desired_dl/spectrum_dl)

    
    print(band, desired_dl, spectrum_dl, Npix)
    
    kernel = astropy.convolution.Box1DKernel(width=Npix+1)
    
    
    figure(figsize=(15,4))
    
    plot(lc, sc,'r-', lw=1)
    plot(lc, convolve(sc, kernel=kernel),'k-', lw=2)

    grid(True)

    title("%s band" % band)
    xlabel("Wavelength [$\AA$]")
    ylabel("Flux [$photons$ for Magellan in a %s s exposure]" % et)
    ylim(-2000, 20000)
    axvline(l)
    axvline(r)
    #axvline(hpps[1]*10000)

figure(figsize=(10,4))
plot(kernel.array)
title("Kernel")
```

## Now play with H band

```python
figure(figsize=(10,5))
grid(True)

ok = (lc>16010) & (lc<16050)
plot(lc[ok], convolve(sc, kernel=kernel)[ok],'k-', lw=2)

legend(["sp", "Maihara"])

a,b = 16028.8, 16032.8
axvline(a)
axvline(b)

R = (a+b)/(2*(b-a))
print("%4.1f" % R)
```
