```python
from pylab import *
import astropy, astropy.io, astropy.convolution
import astropy.io.fits as FF
```

```python
f = FF.open("MF.20130408.47725.fits")
```

```python
f[0]
```

```python
ker = astropy.convolution.Box2DKernel(25)
```

```python
# Say, "the default sans-serif font is COMIC SANS"
matplotlib.rcParams['font.sans-serif'] = "Avenir"
# Then, "ALWAYS use sans-serif fonts"
matplotlib.rcParams['font.family'] = "sans-serif"

matplotlib.rcParams.update({"font.size": 22})

dat = f[0].data
cut = dat [600:800,:]
#cut = dat [1475:1675,:]
sm = astropy.convolution.convolve(cut, ker)


figure(figsize=(10,5))
imshow(np.random.poisson(sm[50:-20,400:900]*2), vmax=8000, vmin=0, cmap='inferno', interpolation='none')
xlabel("Wavelength [pixel]")
title("Simulated Henrietta Spectrum")

savefig("henrietta_2d_targ.pdf")
```
