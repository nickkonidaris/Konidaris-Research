```python
import numpy as np
from pylab import *
```

```python
x = np.random.poisson(lam=1, size=2000)
```

```python
print(hist(np.round(x), align="left", bins=np.arange(-1,16)))
xlabel("Number of Dust Events per Spaxel Per Month")
ylabel("Number of Spaxels Affected")
```

```python
x5 = np.random.poisson(lam=5, size=2000)
hist(np.round(x5), align="left", bins=np.arange(-1,16))
xlabel("Number of Dust Events per Spaxel After 5 Months")
ylabel("Number of Spaxels Affected")
```
